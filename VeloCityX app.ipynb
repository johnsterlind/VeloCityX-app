# Analysis of fan engagement for VeloCityX: Driving engagement and monetization strategies

### Introduction
The VeloCityX app is designed to enhance fan engagement and monetize autonomous racing. This analysis explores how fan challenges, virtual merchandise purchases, and sponsorship interactions contribute to engagement and revenue. Using the provided data, we aim to discover trends and propose new fan challenges that boost both engagement and monetization.

```python
import pandas as pd

# Load the dataset from Downloads folder
file_path = '/Users/johnsterlind/Downloads/2025-VeloCityX-Expanded-Fan-Engagement-Data.csv'
data = pd.read_csv(file_path)

# Display the first few rows to inspect the data
data.head()
```

```python
# Check the structure and info of the dataset
data.info()

# Check for missing values
data.isnull().sum()
```

```python
# Drop non-numeric columns (User ID in this case) for correlation analysis
numeric_data = data.select_dtypes(include=['int64', 'float64'])
```

```python
# Perform the correlation analysis on numeric columns
correlation_matrix = numeric_data.corr()

# Display correlation values for 'Virtual Merchandise Purchases'
correlation_matrix['Virtual Merchandise Purchases'].sort_values(ascending=False)
```

### Insights from Correlation Analysis
- Fan Challenges Completed shows the strongest positive correlation (\(0.159\)) with Virtual Merchandise Purchases.
- Sponsorship Interactions (Ad Clicks) also has a positive correlation (\(0.071\)), though weaker.
- Time on Live 360 and Real-Time Chat Activity show no significant correlation with merchandise purchases.

```python
import matplotlib.pyplot as plt

# Scatter plot for Fan Challenges Completed vs Virtual Merchandise Purchases
plt.scatter(data['Fan Challenges Completed'], data['Virtual Merchandise Purchases'])
plt.title('Fan Challenges Completed vs Virtual Merchandise Purchases')
plt.xlabel('Fan Challenges Completed')
plt.ylabel('Virtual Merchandise Purchases')
plt.show()
```

```python
# Scatter plot for Sponsorship Interactions vs Virtual Merchandise Purchases
plt.scatter(data['Sponsorship Interactions (Ad Clicks)'], data['Virtual Merchandise Purchases'])
plt.title('Sponsorship Interactions vs Virtual Merchandise Purchases')
plt.xlabel('Sponsorship Interactions (Ad Clicks)')
plt.ylabel('Virtual Merchandise Purchases')
plt.show()
```

```python
# Scatter plot for Time on Live 360 vs Virtual Merchandise Purchases
plt.scatter(data['Time on Live 360 (mins)'], data['Virtual Merchandise Purchases'])
plt.title('Time on Live 360 vs Virtual Merchandise Purchases')
plt.xlabel('Time on Live 360 (mins)')
plt.ylabel('Virtual Merchandise Purchases')
plt.show()
```

### Proposed Fan Challenge: "Unlock Merch Madness"
To increase engagement and drive virtual merchandise purchases, we propose the "Unlock Merch Madness" challenge:

- **Objective**: Encourage users to complete more fan challenges to unlock exclusive merchandise.
- **Mechanics**: 
    - Users unlock special virtual merchandise (e.g., limited-edition vehicle skins, team jerseys) after completing a set number of challenges.
    - Bonus points are awarded for users who achieve high predictive accuracy in race outcomes.
    - Users who interact with sponsorship ads more frequently can earn discounts on merchandise purchases.

This challenge leverages the positive correlation between fan challenges and merchandise purchases while incentivizing sponsorship interactions.

### Conclusion
This analysis highlights the importance of fan challenges and sponsorship interactions in driving virtual merchandise purchases for VeloCityX. Based on the data, we propose the "Unlock Merch Madness" challenge to further incentivize user engagement and increase monetization. By combining fan challenges with rewards and sponsorship incentives, VeloCityX can create a more immersive and profitable fan experience.
